# Data Cleaning

```python
df = df.drop(columns=['RowNumber','CustomerId','Surname'])
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ax = sns.countplot(x='Exited',data=df)
plt.title("Churn Distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

79.6% of customers were retained, while 20.4% churned, indicating a relatively low but significant customer attrition rate.

```python
ax= sns.countplot(x='Geography', hue='Exited', data=df)
plt.title('Churn by Geography')
ax.bar_label(ax.containers[0])
```

France recorded the highest number of churned customers (4,203), followed by Spain (2,064) and Germany (1,695).

```python
ax= sns.countplot(x='Gender',hue='Exited',data=df)
plt.title("Churn by Gender")
ax.bar_label(ax.containers[0])
```

Male customers account for a higher number of churn cases (4,558) compared to female customers (3,404).

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[18,30,45,60,100], labels=['18-30','31-45','46-60','60+'])
ax= sns.countplot(x='AgeGroup', hue='Exited', data=df)
plt.title('Churn by Age Group')
ax.bar_label(ax.containers[0])
```

Customers aged 31–45 show the highest churn count (4,988), making them the most vulnerable age segment.

```python
ax=sns.countplot(x='IsActiveMember', hue='Exited', data=df)
plt.title('Churn by Active Membership')
ax.bar_label(ax.containers[0])
```

Active members(1) account for the highest churn count (4,416), suggesting that engagement alone does not guarantee customer retention.

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
```

Age shows the strongest positive correlation with churn (0.29), followed by Complaints (0.28), indicating that older customers and customers with complaints are more likely to leave the bank.
